```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dados de Treinamento
x = [[2,1,0],
     [1,0,2],
     [2,2,0],
     [1,2,0],
     [0,0,2],
     [0,1,2]]
y = [1,0,1,1,0,0]
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(x,y)
```

```python
arvore = export_text(modelo, feature_names=["Prescisao Montagem", "Velocidade da Producao", "Taxa de Retrabalho"])
print(arvore)
```

```python
nova_maquina = [[1,0,2]]
classe = modelo.predict(nova_maquina)
print(classe)
```

```python
plt.figure(figsize=(10,6))
tree.plot_tree(
    modelo,
    feature_names=["Prescisão Montagem", "Velocidade da Produção", "Taxa de Retrabalho"],
    class_names=["Alta Qualidade", "Baixa Qualidade"],
    filled=True,
    rounded=True
)
plt.show()
```
